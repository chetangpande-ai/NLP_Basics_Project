```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from collections import defaultdict

# Download once
nltk.download('punkt')
nltk.download('stopwords')
```

```python
text = """
API testing ensures that backend services work correctly.
Automation testing reduces regression effort and improves efficiency.
Manual testing is essential for usability and exploratory scenarios.
API testing is critical in microservices architecture.
"""

# Step 1: Sentence tokenization
sentences = sent_tokenize(text)
```

```python
# Step 2: Word tokenization
words = word_tokenize(text.lower())
# Step 3: Remove stopwords and non-alphabetic words
stop_words = set(stopwords.words("english"))
filtered_words = [w for w in words if w.isalpha() and w not in stop_words]
```

```python
# Step 4: Word frequency table
word_freq = defaultdict(int)
for word in filtered_words:
    word_freq[word] += 1

# Step 5: Sentence scoring
sentence_scores = defaultdict(int)
for sentence in sentences:
    for word in word_tokenize(sentence.lower()):
        if word in word_freq:
            sentence_scores[sentence] += word_freq[word]

# Step 6: Select top sentences
summary_sentences = sorted(sentence_scores, key=sentence_scores.get, reverse=True)[:2]

summary = " ".join(summary_sentences)
print(summary)
```
