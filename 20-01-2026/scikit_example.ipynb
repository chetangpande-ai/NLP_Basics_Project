```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

```

```python
# Input text
text = """
API testing ensures backend services work correctly.
Automation testing reduces regression effort.
Manual testing helps find usability issues.
API testing is critical for microservices architecture.
"""

# Step 1: Sentence split (simple split for demo)
sentences = [s.strip() for s in text.split(".") if s.strip()]

# Step 2: TF-IDF vectorization
vectorizer = TfidfVectorizer(stop_words="english")
tfidf_matrix = vectorizer.fit_transform(sentences)

# Step 3: Sentence scoring
sentence_scores = np.array(tfidf_matrix.sum(axis=1)).ravel()

# Step 4: Select top N sentences
top_n = 2
top_sentence_indices = sentence_scores.argsort()[-top_n:][::-1]

summary = " ".join([sentences[i] for i in top_sentence_indices])

print(summary)
```
