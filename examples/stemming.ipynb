Lemmatization is an NLP technique that reduces a word to its base or dictionary form, called the lemma. Unlike stemming—which simply chops off word endings—lemmatization uses linguistic rules and vocabulary, so the output is always a meaningful word. It considers the part of speech (POS) of a word to return the correct lemma. For example, “running,” “ran,” “runs” → “run”. Another example: “better” → “good”, because lemmatization understands grammar and irregular forms. This helps improve text analysis, search, and machine learning tasks by treating different forms of a word as one.

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

```

```python

ps = PorterStemmer()
words = ["running", "runner", "runs", "easily", "fairly"]

for w in words:
    print(w, "->", ps.stem(w))
```
