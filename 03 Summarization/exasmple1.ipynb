```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from collections import defaultdict

# Download required data
nltk.download('punkt')
nltk.download('stopwords')
```

```python

text = """
Artificial Intelligence is transforming the software testing industry.
AI helps in test case generation, defect prediction, and test optimization.
Manual testing is time-consuming and error-prone.
AI-based tools improve efficiency and reduce human effort.
Many enterprises are adopting AI for quality assurance.
"""

# Tokenize sentences and words
sentences = sent_tokenize(text)
words = word_tokenize(text.lower())

# Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [w for w in words if w.isalnum() and w not in stop_words]

# Calculate word frequency
word_freq = defaultdict(int)
for word in filtered_words:
    word_freq[word] += 1

# Score sentences
sentence_scores = defaultdict(int)
for sent in sentences:
    for word in word_tokenize(sent.lower()):
        if word in word_freq:
            sentence_scores[sent] += word_freq[word]

# Select top 2 sentences
summary_sentences = sorted(
    sentence_scores,
    key=sentence_scores.get,
    reverse=True
)[:2]

summary = " ".join(summary_sentences)

print("Summary:")
print(summary)
```
