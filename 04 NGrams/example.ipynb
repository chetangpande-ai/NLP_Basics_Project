An N-gram is a contiguous sequence of N words from text.

Unigram (1-gram): machine

Bigram (2-gram): machine learning

Trigram (3-gram): machine learning model

ðŸ“Œ Used in language modeling, text prediction, search, classification, spam detection.

```python
import nltk
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
```

```python
# Download tokenizer
nltk.download('punkt')

text = "Hi, I am  chetan. Currently works as automation solution architect at Northern Trust, Pune."

# Tokenize text
tokens = word_tokenize(text.lower())

# Generate bigrams
bigrams = list(ngrams(tokens, 2))

# Generate trigrams
trigrams = list(ngrams(tokens, 3))

print("Tokens:")
print(tokens)

print("\nBigrams:")
print(bigrams)

print("\nTrigrams:")
print(trigrams)
```
