```python
corpus="""
hello welcome, this is chetan, i am a software engineer, i have 5 years of experience in software development, i have worked on various projects, i am passionate about coding and learning new technologies, i am always eager to take on new challenges and improve my skills, i believe in continuous learning and growth, i am a team player and enjoy collaborating with others to achieve common goals, i am also a problem solver and enjoy finding solutions to complex issues, overall, i am dedicated to my work and strive for excellence in everything i do.
i want to learn NLP and machine learning, i am interested in natural language processing and its applications, i want to understand how machines can understand and process human language, i am excited about the potential of NLP in various fields such as healthcare, finance, and customer service, i am eager to explore different NLP techniques and algorithms, i want to gain hands-on experience with NLP tools and libraries, i am looking forward to applying my knowledge of NLP to real-world problems and making a positive impact in the field of artificial intelligence.
"""
```

```python
print(corpus)
```

```python
##tokenization
##sentence -->paragraph

from nltk.tokenize import sent_tokenize
sentences = sent_tokenize(corpus)
print(sentences)
```

```python
type(sentences)

for sentence in sentences:
    print(sentence)
```

```python
##tokenization
##paragraph --> words
from nltk.tokenize import word_tokenize
wprds = word_tokenize(corpus)
print(wprds)
```

```python
for word in wprds:
    print(word)
```
